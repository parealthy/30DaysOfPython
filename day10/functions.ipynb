```python
def add_all_nums(*numbers):
    total = 0
    for number in numbers:
        if not str(number).isnumeric():
            return "输入有问题。"
        total += number
    return total

add_all_nums(2,3,'abcc',9,2,'a')
```

```python
def solve_quadratic_eqn(a,b,c):
    delta = b**2 - 4*a*c
    x1 = (-b + delta**(1/2))/2*a
    x2 = (-b-delta**(1/2))/2*a
    return (x1,x2)

print(solve_quadratic_eqn(1,-2,-3))
```

```python
def all_unique(lst):
    set_lst = set()
    for item in lst:
        if item in set_lst:
            return False
        set_lst.add(item)
    return True
```

```python
a = [1,2,3,4,5,6,1,2,3]
```

```python
all_unique(a)
```

```python
def all_same_type(lst):
    tp = type(lst[0])
    for i in range(1,len(lst)):
        if type(lst[i])!=tp:
            return False
    return True
```

```python
def the_most_populated_countries(countries_dict):
    populations = {}
    for country in countries_dict:
        populations[country['name']] = country['population']
    return sorted(populations,key=lambda x:populations[x],reverse=True)[:20]
```

```python
def the_most_spoken_languages(countries_dict):
    languages = {}
    for country in countries_dict:
        spoken_languages = country['languages'] 
        for language in spoken_languages:
            if language not in languages:
                languages[language] = 0
            else:
                languages[language] += 1
    return sorted(languages,key=lambda x:languages[x],reverse=True)[:20]
```
