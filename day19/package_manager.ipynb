```python
from collections import defaultdict
words_dict = defaultdict(int)
for word in words:
    words_dict[word] += 1
```

```python
words_dict
```

```python
import heapq
help(heapq)
```

```python
most_freq_words = heapq.nlargest(10,words_dict,key=lambda x:words_dict[x])
```

```python
weight_max = 0
weight_min = 100
for cat in cats_data:
    weight = cat['weight']['metric']
    min_w = int(weight[0])
    max_w = int(weight[-1])
    if max_w>weight_max:
        weight_max = max_w
    if min_w < weight_min:
        weight_min = min_w

print(weight_max,weight_min)
```
