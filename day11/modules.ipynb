```python
def random_use_id(nums):
    alnum = string.ascii_letters + string.digits
    return ''.join(random.choices(alnum,k=int(nums)))
```

```python
def user_id_gen_by_use():
    characters_num = input('请输入字符数量：')
    ids_num = input('需要生成的ID个数：')
    for _ in range(int(ids_num)):
        print(random_use_id(characters_num))
```

```python
def list_of_hexa_colors(nums):
    hexa_colors = []
    hexa_letters = "0123456789abcdef"
    for _ in range(nums):
        hexa_colors.append('#'+''.join(random.choices(hexa_letters,k=6)))
    return hexa_colors
print(list_of_hexa_colors(5))
```

```python
random.choices(range(256),k=3)
```

```python
def list_of_rgb_colors(nums):
    rgb_colors = []
    for _ in range(nums):
        rgb_colors.append('rgb'+str(tuple((random.choices(range(256),k=3)))))
    return rgb_colors

print(list_of_rgb_colors(5))
```

```python
def generate_colors(mode,nums):
    if mode == 'hexa':
        return list_of_hexa_colors(nums)
    else:
        return list_of_rgb_colors(nums)
print(generate_colors('hexa',3))
print(generate_colors('rgb',3))
```

```python
def shuffle_list(lst):
    copied = lst.copy()
    random.shuffle(copied)
    return copied
```

```python
a = ['beijing','shanghai','hangzhou','suzhou','ningbo']
print(a)
print(shuffle_list(a))
print(a)
```
